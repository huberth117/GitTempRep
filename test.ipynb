```python
import pandas as pd
import numpy as np

np.random.seed(42)

# Número de registros
num_reg = 100

# Variables id, edad, genero, producto, cantidad, precio
id = np.arange(1, num_reg + 1)
edad = np.random.randint(18, 180, num_reg)
genero = np.random.choice(['M', 'F', 'I'], num_reg)
producto = np.random.choice(['Silla', 'Mesa', 'Portavaso'], num_reg)
cantidad = np.random.randint(1, 100, num_reg)
precio = np.random.uniform(5, 100, num_reg)

# Conjunto de datos
df = pd.DataFrame({
      'ID' : id,
      'Edad' : edad,
      'Género' : genero,
      'Producto' : producto,
      'Cantidad' : cantidad,
      'Precio' : precio})

df
```

```python
print(df.dtypes)
print(df.sum())
```

```python
df['Total_gasto'] = df['Cantidad'] * df['Precio']

df['Género'] = df['Género'].map({'M':0, 'F':1, 'I':2 })
```
